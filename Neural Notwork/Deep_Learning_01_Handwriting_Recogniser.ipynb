```python
# normal model making
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784, ), activation = "sigmoid")
])
```

```python
model.summary()
```

```python
adam = tf.keras.optimizers.Adam(learning_rate = 0.001)

model.compile(
    optimizer = adam,
    loss = tf.keras.losses.sparse_categorical_crossentropy,
    metrics = ["accuracy"]
    
)

bot = model.fit(x_train_flat, y_train, epochs= 10)
```

```python
model.evaluate(x_test_flat, y_test)
```

### Plotting the loss epoch wise

```python
plt.xlabel('Epoch Number')
plt.ylabel('loss')
plt.plot(bot.history['loss'], label='training set')
plt.legend()
```

### Plot the Accuracy epoch wise

```python
plt.xlabel('Epoch Number')
plt.ylabel('Accuracy')
plt.plot(bot.history['accuracy'], label='training set')
plt.legend()
```

- as the y_test is consist of integer values but the y_pred is consist of whole values, we can convert the whole values as the class lebel(0-9) of each values.

### Visualize the confusion matrix using the heatmap

```python
import seaborn as sns
plt.figure(figsize=(16, 12))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Truth value")
plt.show()
```

- **Enything which are not in the diagonal is the error**

## Adding hidden layer into the Neural Network 

```python
model = keras.Sequential([
    # First layer is the hidden layer
    keras.layers.Dense(100, input_shape = (784, ), activation = "relu"),
    # last layer doesn't need input shape, because whatever 
    # 1st layer is connected it knows input is that
    keras.layers.Dense(10, activation = "sigmoid")
    
])

model.compile(
    optimizer="adam",
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
    
)

nn_bot = model.fit(x_train_flat, y_train, epochs= 10)
```

- **The Accuracy after adding the hidden layer is - $0.9937$**

```python
model.evaluate(x_test_flat, y_test)
```

## Comparison of accuracy with and without hidden layer

```python
plt.xlabel('Epoch Number')
plt.ylabel('Accuracy')
plt.plot(bot.history['accuracy'], label='Without hidden layer')
plt.plot(nn_bot.history['accuracy'], label='with hidden layer')
plt.legend()
```

```python
y_pred = model.predict(x_test_flat)
y_pred_lebel = [np.argmax(i) for i in y_pred]
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_lebel)

import seaborn as sns
plt.figure(figsize=(16, 12))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Truth value")
plt.show()
```
